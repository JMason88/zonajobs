```python
import pandas as pd
import numpy as np
```

```python
submision = pd.read_csv('../salidas/submission.csv')

submision.head()
```

```python
avisosAPredecir = pd.read_pickle('../preprocessing/entrada/avisos_a_predecir.pkl')

print(avisosAPredecir)
```

```python
top25 = {'idaviso': avisosAPredecir, 'ranking': np.array(range(25))}
df_top25 = pd.DataFrame(top25)
df_top25.head()
```

```python
df_merge = submision.merge(
    right = df_top25,
    how = 'left',
    on = 'idaviso'
)

df_merge.head(25)
```

```python
train = pd.read_csv('../data/postulaciones_train.csv')
train = train[:100000]
train.head()
```

```python
trainMerge = train.merge(
    right = df_merge,
    how = 'inner',
    on = ['idaviso', 'idpostulante']
)

trainMerge.head(25)
```
